```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as  sns
%matplotlib  inline
```

Changing Formatted Date from  string to datetime

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'],utc=True)
```

Changing index to Formatted Date

```python
df=df.set_index(["Formatted Date"])
```

```python
df
```

Resampling Formatted Date, Hourly to Monthly starting

```python
df_column=["Apparent Temperature (C)","Humidity"]
df_monthly_mean=df[df_column].resample("MS").mean()
# MS- Monthly starting
```

```python
df_monthly_mean
```

**Null Hypothesis H0:** Has the Apparent Temperature and Humidity compared monthly across 10 years of the data indicate an increase due to global warming

**Alternate Hypothesis H1**: Has the Apparent Temperature and Humidity compared monthly across 10 years of the data not indicate an increase due to global warming

```python
sns.regplot(data=df_monthly_mean,x="Apparent Temperature (C)",y="Humidity")
plt.title("Relation between Apparent Temperature and Humidity",size=22)
plt.grid()
```

From the above graph, we can observe that, there might be Linear Relationship between Apparent Temperature and Humidity with negative slope

```python
sns.lineplot(data=df_monthly_mean)
plt.legend()
plt.title("Variation of Apparent Temperature and Humidity over time",size=22)
plt.xlabel("Formatted Date",size=18)
```

**Observation :**
Apparent Temperature is frequently changed from 2006-2016 and
Humidity remamined constant from 2006-2016

```python
sns.heatmap(data=df_monthly_mean.corr(),annot=True)  #annot=True indicates show labels on heatmap
plt.title("Correlation between Apparent Temperature and Humidity",size=18)
```

Function for plotting Apparent Temperature and Humidity for all months

```python
TEMP_DATA=df_monthly_mean.iloc[:,0]
HUM_DATA=df_monthly_mean.iloc[:,1]
def label_color(month):
  if month==1:
    return 'January','red'
  elif month==2:
    return 'February','blue'
  elif month==3:
    return 'March','green'
  elif month==4:
    return 'April','black'
  elif month==5:
    return 'May','orange'
  elif month==6:
    return 'June','pink'
  elif month==7:
    return 'July','grey'
  elif month==8:
    return 'August','purple'
  elif month==9:
    return 'September','yellow'
  elif month==10:
    return 'October','brown'
  elif month==11:
    return 'November','darkblue'
  else:
    return 'December','violet'

def plot_month(month,data):
  label,color=label_color(month)
  mdata=data[data.index.month==month]
  sns.lineplot(data=mdata,label=label,color=color,marker='o')
  #plt.show()

def sns_plot(title,data):
  plt.figure(figsize=(14,8))
  plt.title(title)
  plt.xlabel("Year")
  for i  in range(1,13):
    plot_month(i,data)
  plt.show()
  
```

```python
# Month wise plot for Apparent Temperature for 10 years
title='Month wise plot for Apparent Temperature for 10 years'
sns_plot(title,TEMP_DATA)
```

```python
# Month wise plot for Humidity for 10 years
title="Month wise plot for Humidity for 10 years"
sns_plot(title,HUM_DATA)
```

Function for plotting Apparent Temperature and Humidity for each month

```python
def sns_month_plot(month):
  plt.figure(figsize=(10,6))
  data=df_monthly_mean[df_monthly_mean.index.month==month]
  sns.lineplot(data=data,marker='o')
  plt.xlabel("Year")
  label=label_color(month)[0]
  plt.title("Apparent Temperature Vs Humidity for {0}".format(label))
```

```python
# Plot for months from  January-December for 10 years
for i in range(1,13):
  sns_month_plot(i)
```

**Conclusion:**

1.  Humidity remained constant over the years 2006-2016.
2.  Apparent Temperature has faced ups and downs over the years 2006-2016
3.  So we can conclude that,global warming has caused an uncertainty in the Apparent Temperature over the 10 years.
